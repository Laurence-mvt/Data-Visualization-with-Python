```python
# read in the data 
df_can = pd.read_excel('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx',
sheet_name='Canada by Citizenship',
skiprows=range(20),
skipfooter=2)

print('Data read into a pandas dataframe!')
```

Clean data set by removing some unnecessary columns

```python
# note in pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
# drop emppty columns 
df_can.drop(df_can.columns[-8:-1], axis=1, inplace=True)
df_can.head(2)
```

```python
# rename columns to make sense
df_can.rename(columns={'OdName': 'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

```python
# add a total column
df_can['Total'] = df_can.sum(axis=1)
```

```python
# how many null objects in dataset
df_can.isnull().sum()
```

```python
df_can.describe()
```

```python
df_can.shape
```

Indexing and slicing with pandas

```python
# select column
# method 1 - only works if column name doesn't have spaces or special characters
df_can.Country
```

```python
# method 2 - preferable, and can select multiple columns
df_can[['Country', 'Continent', 1980, 1981]] 
# note that dates are numbers, not strings. Will convert them to strings soon for consistency
```

```python
# Select rows
# note that the default index for rows is 0,...number of rows -1. 
# first set index to Country:
df_can.set_index('Country', inplace=True)
# note can reset index, using df.reset_index() method
df_can.head(2)
```

```python
df_can.head(3) 
# now 'Country' not column head, but index
```

```python
# select number of immigrants from Japan (row 87)
# the full row 
print(df_can.loc['Japan'])

# alternative methods
print(df_can.iloc[87])
print(df_can[df_can.index == 'Japan'].T.squeeze())
```

```python
# to select year 2013 only
df_can.loc['Japan', 2013]

# alternative methods
df_can.iloc[87, 36]

```

```python
# for years 1980 to 1985
df_can.loc['Japan', [1980, 1981, 1982, 1983, 1984]]
df_can.iloc[87, [3, 4, 5, 6 , 7, 8]]
```

Change the columns to all strings, to avoid confusion

```python
df_can.columns = list(map(str, df_can.columns))
[print (type(x)) for x in df_can.columns.values] #<-- uncomment to check type of column headers
```

```python
# to conveniently call upon full range of years
years = list(map(str, range(1980, 2014)))
years
```

Can filter dataframe based on condition

```python
# eg. 
# 1. Create the condition boolean series
condition = df_can['Continent'] == 'Asia'
print(condition)
```

```python
# 2. Pass this condition into the dataFrame
df_can[condition]
```

```python
# can pass multiple criteria
df_can[(df_can['Continent'] == 'Asia') &  (df_can['Region'] == 'Southern Asia')]
# note pandas requires use of & and | operators instead of 'and' and 'or'
```

```python
# review dataframe
print(df_can.head(5))
print(df_can.info(verbose=False))
print('data dimensions', df_can.shape)
```

Visualization with Matplotlib

```python
# we are using the inline backend
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
print('Matplotlib version', mpl.__version__)
```

```python
# add a style to Matplotlib
print(plt.style.available)
# mpl.style.use(['ggplot']) # optional: for ggplot-like style
```

```python
# get line plot of immigration from Haiti
haiti = df_can.loc['Haiti', years]
haiti.head(10)

haiti.index = haiti.index.map(int) # let's change the index values of Haiti to type integer for plotting
haiti.plot(kind='line')
```

```python
# plot
haiti.plot()
```

```python
# label axes
haiti.plot()
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```

```python
# annotate the plot to show earthquake of 2010
haiti.plot(kind='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

# annotate the 2010 Earthquake. 
# syntax: plt.text(x, y, label)
plt.text(2000, 6000, '2010 Earthquake')

plt.show()
```

Compare migration from China and India


```python
India = df_can.loc['India', years]
China = df_can.loc['China', years]
```

```python
print(India.head())
```

```python
# plot graph
China.plot(kind='line')
India.plot(kind='line')
plt.title('Immigration from China and India')
plt.ylabel('Immigration')
plt.xlabel('Years')
```

```python
# alternatively, make df for both China and India
df_CI = df_can.loc[['China', 'India'], years]
```

```python
df_CI.head()
```

```python
df_CI.plot(kind='line')
# not pandas plots the indices on the x-axis and columns as individual lines on the y-axis, so first need to transpose dataframe using transpose() method 
df_CI = df_CI.transpose()
df_CI.head()

df_CI.plot(kind='line')
plt.title('Immigration from China and India')
plt.ylabel('Immigration')
plt.xlabel('Years')
```

```python
# Compare the trend of top 5 countries that contributed the most to immigration to Canada.
# sort df by total immigration
df_can.sort_values('Total', ascending=False, axis=0, inplace=True)
# get top 5 countries
df_top5 = df_can.iloc[:5][years]
df_top5.head()
# transpose
df_top5 = df_top5.transpose()
df_top5.head()
df_top5.plot(kind='line')

plt.title("Immigration to Canada by top 5 countries")
plt.ylabel('Immigration')
plt.xlabel('Years')

plt.show()
```
